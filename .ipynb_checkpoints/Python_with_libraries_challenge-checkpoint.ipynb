# Python Programming with Libraries 

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy
from scipy import stats
import statistics

import seaborn as sns
sns.set()
```

### Read CSV data from two files and merge it into a single Data Frame 

```python
students_df = pd.read_csv('task2a.csv')
click_events_df = pd.read_csv('task2b.csv')

merged_df = pd.merge(students_df, click_events_df)
```

### Clean the merged data

```python
#remove rows with missing data in any column 
merged_df.dropna(inplace=True)

# removed unnecessary columns
cleaned_df = merged_df.drop(labels=['Unnamed: 0', 'region', 'final_result', 'highest_education'], axis=1, inplace=False)
# print(cleaned_df.head())

# also removed unnamed: 0 as the same as the index (first column)
```

###  Filter out unnecessary rows

```python
# remove rows where click evs > 20000 

cleaned_df = cleaned_df[cleaned_df.click_events <= 20000]

# remove duplicate rows based on id_student 

cleaned_df.drop_duplicates(subset='id_student', inplace=True)
cleaned_df.head()
```

### Investigate the effects of engagement on attainment

```python
# sample of data
#sample = cleaned_df.sample(n=500, random_state=50)
#x = sample.click_events
#y = sample.score 

# use whole data to show relationship between engagement and the score attained 
x = cleaned_df.click_events 
y = cleaned_df.score

# create scatter plot
plt.scatter(x, y, marker='.', color='orange')
plt.xlabel('Number of click events')
plt.ylabel('Score attained (%)')
plt.title('Number of click events vs the score attained by the student')

# perform simple linear regression to find the regression line for the sample and plot onto graph
regression = scipy.stats.linregress(x, y)
print(regression)
slope = regression.slope
intercept = regression.intercept

# plot the function for the regression line 
line_function = 'y = {}x + {}'.format(round(slope, 4), round(intercept, 2))
plt.plot(x, intercept + (slope*x), label = line_function )
plt.legend()

```

The regression line shows that as the number of click events increases by 1, the score attained increases by 0.0022. 

```python
# group the scores, calculate mean for each group
score_group = pd.cut(cleaned_df.score, bins=[0,25,50,75,100], labels=['<25', '25-50', '50-75','75-100'])
cleaned_df.insert(6, 'score_group', score_group)
```

```python
# produce bar chart displaying mean number of click events for each score group
avg_clicks_per_score = cleaned_df.groupby('score_group').click_events.mean().reset_index()
print(avg_clicks_per_score.head())
score_clicks_plt = sns.barplot(avg_clicks_per_score, x='score_group', y='click_events')


# loop through each bar of the bar chart to display the mean number of click events above the each bar 
# round the 0 dp and center the value 
for i, b in enumerate(score_clicks_plt.patches):
    score_clicks_plt.annotate(f'{b.get_height():,.0f}', xy = (b.get_x() + b.get_width() / 2, b.get_height()), ha= 'center')


plt.title('The mean number of click events per scores grouped into four quarters')
plt.xlabel('Score (%)')
plt.ylabel('Mean number of click events')
plt.show()
```

These graphs show a relationship between the number of click events and the score. 

The mean number of click events increases with each each group of scores. There is a postive trend in the data - as the number of click events increases, the score also increases (by 0.0022 according to the regression line on the scatter graph). This suggests that the more a student engages (clicks) on the online material, the better their score will be.



### Test the hypothesis that engagement has some effect on levels of attainment 

```python
# check the distribution of the variables by plotting histograms
fig, ax = plt.subplots(2)
ax[0].hist(x)
ax[0].set_title('Histogram of number of click events')
ax[1].hist(y)
ax[1].set_title('Histogram of score')
fig.subplots_adjust(hspace=0.5)
```

The two variables appear to not be normally distributed as they show skew and the variable score shows an outlier. Therefore Pearson correlation should not be used as that requires normally distributed data. Spearman correlation can be used instead.

```python
# null hypothesis: engagement has no effect on levels of attainment 


# perform corrolation on whole dataset: get Spearmans correlation coefficient and p value

alpha = 0.05
x = cleaned_df.click_events
y = cleaned_df.score 

# calculate correlation coefficient
sr = scipy.stats.spearmanr(x, y)

print('r: {}, p-value: {}'.format(sr.correlation, sr.pvalue))


# reject or do not reject null hypothesis?
if sr.pvalue < alpha:
    print('Reject null hypothesis')
else:
    print('Fail to reject null hypothesis')

    
# research to pick a more appropriate significance level? 
# do anvo between the groups?
```

The null hypothesis: engagement has no effect on the level of attainment. 
The alternative hypothesis: engagement has an effect on the level of attainment. 

The Spearman correlation coefficient for this data is 0.321. This shows a small positive correlation between engagement and level of attainment. A significance level of 5% (alpha = 0.05) was used. 
The statistical test returned a p-value of 0.0. As the p-value is less than 0.05, the null hypothesis that engagement has no effect on the level of attainment is rejected. Therefore, from this data, it looks as if the number of click events has some effect on the score.

### Investigate the effects of gender on levels of attainment 

```python
# create variables for the score of the genders
f = cleaned_df[cleaned_df.gender == 'F']
m = cleaned_df[cleaned_df.gender == 'M']

# print medians and means 
print(f'Female: median {f.score.median()}, mean {round(f.score.mean(),1)}')
print(f'Male: median {m.score.median()}, mean {round(m.score.mean(),1)}')

# put these in a list
genders = {'Female': f.score, 'Male': m.score}
#labs = ['Female', 'Male']

# create figure
fig, ax = plt.subplots(1, figsize=(8,5))

# plot the scores as box plots
ax.boxplot(genders.values(), vert=False, patch_artist=True)
ax.set_yticklabels(genders.keys())
ax.set_xlabel('Score')
```

```python
# use overlaying histograms to view the distributions
sns.histplot(data=m, x='score', bins=20, label='Males', color='pink')
sns.histplot(data=f, x='score', bins=20, label='Females', color='black')
plt.legend()
```

The box plots and the histograms show the distributions, medians and quartiles for the data split by gender. There appears to be very little difference in the medians and quartile values of the two groups. The distributions both show negative skew from the histograms and both have a large number of outliers at the lower score end shown in the box plots. The means, printed above the box plots, show only a 0.1 difference between the scores of males and females.

### Test if there is any difference between the attainment of male and female students

```python
# two sample t test to test if significant difference in the means
# null hypothesis: the means of the attainment are equal between male and female students 
# Significance level - alpha
alpha = 0.05

score_female = f.score
score_male = m.score

# means of the samples
print(round(statistics.mean(score_female), 2))
print(round(statistics.mean(score_male), 2))

# perform two sample t test
ttest_result = scipy.stats.ttest_ind(score_female, score_male)
print(ttest_result)

if ttest_result.pvalue < alpha:
    print('Reject null hypothesis')
else:
    print('Fail to reject null hypothesis')
```

A two-sample t-test was performed on the scores of males and the scores of females. 
The null hypothesis: the means of the attainment are equal between male and female students. 
The test returned a p value of 0.5069. A significance level of 5% (alpha = 0.05). 
As the p value is greater than 0.05 - fail to reject the null hypothesis. 

This coincides with the visualisations above - these show very small differences in the mean, median and distributions of the groups. 

From the results of the statistical test and the visualisations, we cannot say that there is a significant difference between the scores of males and females.
